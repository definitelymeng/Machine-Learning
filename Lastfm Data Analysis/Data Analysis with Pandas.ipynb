# Analyzing Lastfm User Data
In this project, I used Pandas library to analyze user data fron Lastfm to explore a wide range of problems including artists popularities and user lisenting habits.

```python
from pandas import Series, DataFrame
import pandas as pd
from numpy.random import randn
import numpy as np

#create artists dataframe
artists_df = pd.read_csv('artists.dat', encoding="utf-8",sep="\t", index_col='id')
#create multi index dataframe of user_artist
user_artists_df = pd.read_csv('user_artists.dat',encoding="utf-8",sep="\t",index_col=['userID', 'artistID'])

#print exclam
exclam =''
for i in range(40):
    exclam=exclam+'!'
    
#print helper function
def format_print(df):
    print()
    print(exclam)
    print()
    i=0
    for n in df.weight.index:
        if(i>=10):
            break
        print(artists_df.name[n]+'('+ str(n) +')'+' '+str(df.weight[n]))
        i+=1
```

```python
#calculate play counts of artists by summarize on level of artistID
artists_playcounts_df=user_artists_df.sum(level=1).sort_values(by='weight',ascending=False)

print('1. Who are the top artists in terms of play counts? ')
format_print(artists_playcounts_df)
```

```python
#count listener number of artists
listeners_df = user_artists_df.count(level=1).sort_values(by='weight',ascending=False)

print('2. What artists have the most listeners? ')
format_print(listeners_df)
```

```python
#calculate total playcounts of each user
users_playcounts_df = user_artists_df.sum(level=0).sort_values(by='weight',ascending=False)

print('3. Who are the top users in terms of play counts?')
format_print(users_playcounts_df)
```

```python
#calculate average play by divide playcounts dataframe by listeners dataframe
averageplay_df = artists_playcounts_df/listeners_df
#sort average play
averageplay_df = averageplay_df.sort_values(by='weight',ascending=False)

print('4. What artists have the highest average number of plays per listener?')
format_print(averageplay_df)
```

```python
#created new listener dataframe with artists who have no less than 50 listeners
new_listeners_df= listeners_df[user_artists_df.count(level=1)>=50]
#calculate new average play
new_averageplay_df = artists_playcounts_df/new_listeners_df
new_averageplay_df = new_averageplay_df.sort_values(by='weight',ascending=False)

print('5. What artists with at least 50 listeners have the highest average number of plays per listener?')
format_print(new_averageplay_df)
```

```python
#create user_friends dataframe
user_friends_df = pd.read_table('user_friends.dat', encoding="utf-8",sep="\t", index_col='userID')
#create new dataframe recording each user's friends count
friend_num_df = user_friends_df.groupby('userID').count()
#merge dataframe of user play counts and friends counts
new_user_df=pd.merge(users_playcounts_df, friend_num_df , left_index=True, right_on='userID')
#calculate average play for user with 5 or more friends
user_with_more_friends =new_user_df[new_user_df.friendID>=5].weight.sum()/len(new_user_df[new_user_df.friendID>=5])
#calculate average play for user less than 5 friends
user_with_less_friends =new_user_df[new_user_df.friendID<5].weight.sum()/len(new_user_df[new_user_df.friendID<5])


print()
print(exclam)
print()
print('6. Do users with five or more friends listen to more songs?')
print('Average play counts for user with five or more friends: '+ str(user_with_more_friends))
print('Average play counts for user less than five friends: '+str(user_with_less_friends))
print('so we can conclude that users with five or more friends listen to more songs')
```

```python
# create new dataframe with only artistID as index
new_df=user_artists_df.reset_index().set_index('artistID')

# create a function to generate listeners as sets for two artists
# find intersection and union for two sets and calculate similarity
def artist_sim(aid1,aid2):
    set1 = set(new_df.loc[aid1].userID)
    set2 = set(new_df.loc[aid2].userID)
    intersect_sets = set1.intersection(set2)
    union_sets =set1.union(set2)
    print(artists_df.name[aid1]+', '+artists_df.name[aid2]+ ' : '+str(len(intersect_sets)/len(union_sets)))

print()
print(exclam)
print()
print('7.  How similar are two artists?')
artist_sim(735,562)
artist_sim(735,89)
artist_sim(735,289)
artist_sim(89,289)
artist_sim(89,67)
artist_sim(67,735)
```

```python
tagged_df =pd.read_table('user_taggedartists.dat', encoding="utf-8",sep="\t", index_col=['month','year','artistID'])
#dataframe recording total tag number for each artist
tag_total=tagged_df.tagID.count(level=2).sort_values(ascending=False)
#dataframe recording total tag number for each artist in each month
tag_count = tagged_df.groupby(level=[0,1,2]).tagID.count()
#sort tag number for each (month,year) in descending order
top_tag=tag_count.sort_values(ascending=False).sort_index(level=[0,1], sort_remaining=False)


#reset index and extract ordered year and month as list
tagged_df=tagged_df.reset_index().set_index(['month','year']).sort_index(level = ['year','month'])
timelist=[]
for (month,year) in tagged_df.index:
    if (month,year) not in timelist:
        timelist.append((month,year))
    else:
        continue

# traversal through month and year to find the fisrt month in top 10 for certain artist
def first_month(aid,df,mylist):
    for (month,year) in mylist:
        if (aid in df[month][year].index[:10]):
            print ('first month in top10 = '+str(month)+' '+str(year))
            return

# traversal through month and year to find total months in top 10 for certain artist
def total_month(aid,df,mylist):
    i=0
    for (month,year) in mylist:
        if (aid in df[month][year].index[:10]):
            i+=1
    return i   

# for artists in top 10 of total tag number, find their total number of tags,
# first month in top 10 and total month in top 10
for aid in tag_total.index[:10]:
    print(artists_df.name[aid]+'('+ str(aid) +')'+': num tags = '+str(tag_total[aid]))
    first_month(aid,top_tag,timelist)
    totaltime= total_month(aid,top_tag,timelist)
    print('months in top10 = '+ str(totaltime))
    print()
    i+=1
```
