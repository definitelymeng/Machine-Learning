## Exploring the Data

```python
# Import libraries necessary for this project
import numpy as np
import pandas as pd
from time import time
from IPython.display import display 
import seaborn as sns
import matplotlib.pyplot as plt

# import warnings filter
from warnings import simplefilter
# ignore all future warnings
simplefilter(action='ignore', category=FutureWarning)

%matplotlib inline

# Load the Census dataset
data = pd.read_csv("census.csv")

# Display the top five record
display(data.head(n=5))
```

### Understand the data

```python
# Total number of records
n_records = len(data.index)

# Number of records where individual's income is more than $50,000
n_greater_50k = np.sum(data['income']=='>50K')

# Number of records where individual's income is at most $50,000
n_at_most_50k = np.sum(data['income']=='<=50K')

# Percentage of individuals whose income is more than $50,000
greater_percent = n_greater_50k/(n_greater_50k+n_at_most_50k)*100

# Print the results
print("Total number of records: {}".format(n_records))
print("Individuals making more than $50,000: {}".format(n_greater_50k))
print("Individuals making at most $50,000: {}".format(n_at_most_50k))
print("Percentage of individuals making more than $50,000: {}%".format(greater_percent))
```

```python
# using pairplot to see distributions of continuous variables and relationships between them according to income
sns.pairplot(data,hue='income')
```

----
## Data Preprocessing

### Transforming Skewed Continuous Features

```python
# get summary statistics of numerical features
data.describe()
```

```python
# from above we can see capital gain and capital loss has very high standard deviation
# plot these two features to see their distribution
def plot_distribution(records): 
    fig = plt.figure(figsize = (8,5))
    ax = fig.add_subplot(1,1,1)
    ax.hist(records,bins=25,color='orange')
    ax.set_title("'%s' Feature Distribution"%(records.name), fontsize = 14)
    ax.set_xlabel(records.name)
    ax.set_ylabel("Frequency")
    ax.set_ylim((0, 2000))
    ax.set_yticks([0, 500, 1000, 1500, 2000])
    ax.set_yticklabels([0, 500, 1000, 1500, ">2000"])
plot_distribution(data['capital-gain'])
plot_distribution(data['capital-loss'])
```

```python
## from plots we can see the distributions are right skewed with extreme outliers for each feature. These extreme
## values can affect model performance, so we can use logarithmix transformation to reduce the influence of
## very large and very small values

#first split data into features and target
y = data['income']
X = data.drop('income', axis=1)
```

```python
# Conduct log-transform
X[['capital-gain','capital-loss']] = X[['capital-gain','capital-loss']].apply(lambda x: np.log(x + 1))

# Visualize the new log distributions
plot_distribution(X['capital-gain'])
plot_distribution(X['capital-loss'])
```

### Normalizing Numerical Features

```python
# Import sklearn.preprocessing Scaler
from sklearn.preprocessing import MinMaxScaler

# Initialize a scaler, then apply it to the features
scaler = MinMaxScaler() 
numerical = ['age', 'education-num', 'capital-gain', 'capital-loss', 'hours-per-week']
X[numerical] = scaler.fit_transform(X[numerical])

# display top five rows of features
display(X.head(n = 5))
```

### Encode categorical variables

```python
# One-hot encode the features
X = pd.get_dummies(X)

# Encode the target variable
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)

# Print the number of features after one-hot encoding
encoded = list(X.columns)
print("{} total features after one-hot encoding.".format(len(encoded)))

# Uncomment the following line to see the encoded feature names
print(encoded)
```

### Create training and testing set

```python
# Import train_test_split
from sklearn.model_selection import train_test_split

# Split the 'features' and 'income' data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    test_size = 0.2, 
                                                    random_state = 0)

# Show the results of the split
print("Training set has {} samples.".format(X_train.shape[0]))
print("Testing set has {} samples.".format(X_test.shape[0]))
```

----
## Building Model

### Naive Predictor
Build a naive predictor which always predict >=50k and evaluate its performance

```python
# Calculate accuracy, precision and recall
accuracy = np.sum(y)/y.size
recall = 1
precision = np.sum(y)/y.size

# Calculate F-score with beta = 0.5
fscore = (1+0.25)*precision*recall/(0.25*precision+recall)

# Print the results 
print("Naive Predictor: [Accuracy score: {:.4f}, F-score: {:.4f}]".format(accuracy, fscore))
```

###  Supervised Learning Models

### 1. Creating a Training and Predicting Pipeline

```python
# Import fbeta_score and accuracy_score from sklearn
from sklearn.metrics import fbeta_score, accuracy_score
def train_predict(learner, X_train, y_train, X_test, y_test): 
    '''
    inputs:
       - learner: the learning algorithm to be trained and predicted on
       - X_train: features training set
       - y_train: income training set
       - X_test: features testing set
       - y_test: income testing set
    '''
    
    results = {}
    
    start = time() # Get start time 
    learner = learner.fit(X_train,y_train)
    end = time() # Get end time
    
    #Calculate the training time
    results['train_time'] = end-start
    
    #predict on test set
    predictions_test = learner.predict(X_test)
        
    # Compute accuracy on test set 
    results['acc_test'] =accuracy_score(y_test,predictions_test)
         
    # Compute F-score on the test set
    results['f_test'] = fbeta_score(y_test,predictions_test,beta=0.5)
       
    print("{} trained in {} seconds.".format(learner.__class__.__name__,results['train_time']))
        
    # Return the results
    return results
```

### 2.  Build Model

```python
# import classifiers
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.linear_model import LogisticRegression\

# Initialize the three models
clf_A = RandomForestClassifier(random_state=14)
clf_B = AdaBoostClassifier(random_state=14)
clf_C = LogisticRegression(random_state=14)

# Collect results on the learners
results = {}
for clf in [clf_A, clf_B, clf_C]:
    clf_name = clf.__class__.__name__
    results[clf_name] = train_predict(clf, X_train, y_train, X_test, y_test)

# # Run metrics visualization for the three supervised learning models chosen
# vs.evaluate(results, accuracy, fscore)
```

```python
#build dictionary for each metrics for visualization
train_time = {}
acc_test ={}
f_test ={}
for key,value in results.items():
    train_time[key]=value['train_time']
    acc_test[key]=value['acc_test']
    f_test[key]=value['f_test']
```

```python
# create bar plot
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
axs[0].bar(list(train_time.keys()), list(train_time.values()))
axs[0].set_title('Train time')
axs[1].set_title('Accuracy')
axs[2].set_title('F-score')
axs[1].bar(list(acc_test.keys()), list(acc_test.values()))
axs[2].bar(list(f_test.keys()), list(f_test.values()))
fig.autofmt_xdate()
```

----
## Model Tuning

From the metrics above, we may conclude that Adaboost is the most appropriate model. Even though it takes longer time to train ompared to other two models, it shows highest F-score for testing data.
Adaboost is sensitive to outliers, as outliers will force the algorithm to fit cases that are unrealistic. For numberical features, scaling helped reduce the effect of outlier and extreme values, so it is suitable for Adaboosting.
We can use Grid Search to imporve our AdaBoost model

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer

clf = AdaBoostClassifier(random_state=14)

# Create the parameters list we wish to tune
parameters = {'n_estimators':[10,50,100,200],'learning_rate':[0.1,1,5]}

# Make an fbeta_score scoring object using make_scorer()
scorer =make_scorer(fbeta_score,beta=0.5)

# Perform grid search on the classifier using 'scorer' as the scoring method
grid_obj = GridSearchCV(clf,parameters,scoring=scorer)

# Fit the grid search object to the training data
grid_fit = grid_obj.fit(X_train,y_train)

# Get the estimator
best_clf = grid_fit.best_estimator_

# Make predictions using the unoptimized and optimized model
predictions = (clf.fit(X_train, y_train)).predict(X_test)
best_predictions = best_clf.predict(X_test)

# Report the before-and-afterscores
print("Unoptimized model\n------")
print("Accuracy score on testing data: {:.4f}".format(accuracy_score(y_test, predictions)))
print("F-score on testing data: {:.4f}".format(fbeta_score(y_test, predictions, beta = 0.5)))
print("\nOptimized Model\n------")
print("Final accuracy score on the testing data: {:.4f}".format(accuracy_score(y_test, best_predictions)))
print("Final F-score on the testing data: {:.4f}".format(fbeta_score(y_test, best_predictions, beta = 0.5)))
```

### Final Model Evaluation

The scores of optimized model are better than that of unoptimized ones, with small increase in acuuracy and F-score. It also performs much better than naive predictor, with over 100% improvement in accuracy and F-score.

## Feature Importance

Find out which features provide the most predictive power. 

```python
#extract feature importances from model
importances = best_clf.feature_importances_ 

# Plot importance
def feature_plot(importances, X_train):   
    # Display the five most important features
    indices = np.argsort(importances)[::-1]
    columns = X_train.columns.values[indices[:5]]
    values = importances[indices][:5]
    # Creat the plot
    fig = plt.figure(figsize = (9,5))
    plt.title("Normalized Weights for First Five Most Predictive Features", fontsize = 16)
    plt.bar(np.arange(5), values, width = 0.6, align="center", color = '#00A000', \
          label = "Feature Weight")
    plt.bar(np.arange(5) - 0.3, np.cumsum(values), width = 0.2, align = "center", color = '#00A0A0', \
          label = "Cumulative Feature Weight")
    plt.xticks(np.arange(5), columns)
    plt.xlim((-0.5, 4.5))
    plt.ylabel("Weight", fontsize = 12)
    plt.xlabel("Feature", fontsize = 12)
    
    plt.legend(loc = 'upper center')
    plt.tight_layout()
    plt.show() 
    
feature_plot(importances, X_train)
```

We can see that top 5 'important' features are numerical features, because they provide more opportunities for optimal splits of the data.

### Feature Selection
Investigate how does a model perform if only a subset of all the available features are used. With less features required to train, the expectation is that training and prediction time is much lower — at the cost of performance metrics. From the visualization above, we see that the top five most important features contribute more than half of the importance of all features present in the data

```python
# Import functionality for cloning a model
from sklearn.base import clone

# Reduce the feature space
X_train_reduced = X_train[X_train.columns.values[(np.argsort(importances)[::-1])[:5]]]
X_test_reduced = X_test[X_test.columns.values[(np.argsort(importances)[::-1])[:5]]]

# Train on the "best" model found from grid search earlier
clf = (clone(best_clf)).fit(X_train_reduced, y_train)

# Make new predictions
reduced_predictions = clf.predict(X_test_reduced)

# Report scores from the final model using both versions of data
print("Final Model trained on full data\n------")
print("Accuracy on testing data: {:.4f}".format(accuracy_score(y_test, best_predictions)))
print("F-score on testing data: {:.4f}".format(fbeta_score(y_test, best_predictions, beta = 0.5)))
print("\nFinal Model trained on reduced data\n------")
print("Accuracy on testing data: {:.4f}".format(accuracy_score(y_test, reduced_predictions)))
print("F-score on testing data: {:.4f}".format(fbeta_score(y_test, reduced_predictions, beta = 0.5)))
```

The final model' F-score and accuracy score is slightly smaller than original ones, with a 3.1% decrease in accuracy and 6.7% decrease in F-score. If the training time of reduced data is siginificantly less than using full dataset, we can consider using reduced data, but it also depends on what kind of problem we are facing. If the problem requires higher F-score, we need to scarifice training time.
